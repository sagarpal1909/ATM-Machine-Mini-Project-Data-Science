```python
import random
import sys
from getpass import getpass
class ATM():
    def __init__(self):
        self.balance = 0
        self.check_bal=1000000.0
         
    def deposit(self, amount):
        self.amount = amount
        self.balance = self.balance + self.amount
        print("Current account balance: Nu.", self.balance)
        print()
 
    def withdraw(self, amount):
        self.amount = amount
        if self.amount > self.balance:
            print("Insufficient fund!")
            print(f"Your balance is Nu.{self.balance} only.")
            print("Try with lesser amount than balance.")
            print()
        else:
            self.balance = self.balance - self.amount
            print(f"Nu.{amount} withdrawal successful!")
            print("Current account balance: Nu.", self.balance)
            print()
 
    def check_balance(self):
        print("Available balance: Nu.", self.balance)
        print()
    
    
    def transfer_to_savings(self, amount):
        check_bal=self.check_bal-amount
        self.balance=self.balance+amount
        print(f"Your money has been Successfully transfered to your savings balance from your checking balance/n The new balance is :: /n Checking Balance={self.check_bal}   | |   Savings Balance={self.balance}")
 
    def transaction(self):
        print("""
            TRANSACTION 
        *********************
            Menu:
            1. Check your savings balance
            2. Deposit Funds
            3. Widraw Funds
            4. Transfer Funds from checking balance to savings balance
            5. Exit
        *********************
        """)
        
        


        

        
print("*******WECOME TO ATM-SERVICE*******")
print("___________________________________________________________________________________________________________\n")
print("----------KINDLY   LOGIN----------")
atm = ATM()        
        

while True:
    pin=getpass(" Enter your 4 digit pin : : " )
    print("re-enter your pin")
    pin_ver=input()
    if(pin==pin_ver):   #PIN VERIFICATION
        print("You can continue")
        break
    else:
        print("PINs does not match. RE-ENTER YOUR PIN")
            
while True:            
    print(" Kindly choose your option from following drop down ::  \n\n 1. check your savings balance \n 2.Deposit money \n 3. Widraw money \n 4.Transfer amount from checking balance to savings balance \n 5.EXIT")
    try:
        option = int(input("Enter 1, 2, 3, 4 or 5:"))
    except:
        print("Error: Enter 1, 2, 3, 4, or 5 only!\n")
        continue
    else:
            if option == 1:
                atm.check_balance()
            elif option == 2:
                amount = float(input("How much you want to deposit(Nu.) :: "))
                atm.deposit(amount)
            elif option == 3:
                amount = float(input("How much you want to withdraw(Nu.) :: "))
                atm.withdraw(amount)
            elif option == 4:
                print(f" Your checking Balance is currently {1000000.0}")
                amount = float(input("Enter the amount you want to transfer :: "))
                atm.transfer_to_savings(amount)
                   
            elif option == 5:
                print(f"""
                    TRANSACTION RECEIPT 
          ******************************************
              Transaction is now complete.                         
              Transaction number: {random.randint(10000, 1000000)}            
              Available balance: Nu.{atm.check_balance()}                    
 
            Thanks for choosing our ATM service                  
          ******************************************
          """)
                sys.exit()
```
